# STATSAPI MLB BASEBALL DATA

```python
import statsapi
import pandas as pd
```

```python
sched = statsapi.schedule(start_date='07/01/2018',end_date='07/31/2018',team=143 ,opponent=121)
```

```python
# API CALL TO RETREIVE JSON FILE
sched[0]
```

```python
# PANDAS DATAFRAME CONVERTS JSON TO TABLE
pd.DataFrame(sched)
```

```python
# PRINTS ROSTER AS OF DATE EXAMPLE
# https://github.com/toddrob99/MLB-StatsAPI/wiki/Function:-get (LINK FOR statsapi.get FUNCTION)
# https://github.com/toddrob99/MLB-StatsAPI/wiki/Function:-roster (LINK FOR statsapi.roster FUNCTION)


print('Phillies 40-man roster on opening day of the 2018 season:\n%s' % 
      statsapi.roster(143,'40Man',date=
                      statsapi.get('season',{'seasonId':2018,'sportId':1})['seasons'][0]['regularSeasonStartDate']))
```

```python
# Final Line Score

for x in [y for y in statsapi.schedule(team=143,start_date='07/01/2008',end_date='07/15/2008') if y.get('winning_team','')=='Philadelphia Phillies']:
    print('%s\nWinner: %s, Loser: %s\n%s\n\n' % (x['game_date'], x['winning_team'], x['losing_team'], statsapi.linescore(x['game_id'])))
```

```python
# PRINT STANDINGS AS OF DATE
# https://github.com/toddrob99/MLB-StatsAPI/wiki/Examples#print-the-standings-from-july-4-2018

print(statsapi.standings(date='08/15/2021'))
```

```python
# GET STATS FOR A PLAYER
# https://github.com/toddrob99/MLB-StatsAPI/wiki/Function:-player_stats

print( statsapi.player_stats(next(x['id'] for x in statsapi.get('sports_players',{'season':2008,'gameType':'W'})['people'] if x['fullName']=='Chase Utley'), 'hitting', 'career') )
```
